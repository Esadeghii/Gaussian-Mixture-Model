```python
import matplotlib.pyplot as plt
```

```python
#latent 25, lstm 15
gamma =[0.01, 0.08, 0.5, 0.7, 1, 1.3, 1.7, 2]
tacc=[0.994,0.962,0.923, 0.931,0.922,0.844, 0.827, 0.830]
vacc=[0.988,0.892,0.829, 0.813,0.823,0.732, 0.688, 0.676]

plt.xlabel("gamma")
plt.ylabel("accuracy")

x= plt.plot(gamma, tacc, color = 'green',
         linestyle = 'solid', marker = 'o',
         markerfacecolor = 'red', markersize = 12, label='training')
y= plt.plot(gamma, vacc, color = 'red'
        ,linestyle='dashed', marker = 'o',
         markerfacecolor = 'blue', markersize = 12, label= 'validation')

plt.legend(loc="upper right")
```
