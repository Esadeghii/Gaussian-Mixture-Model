```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


def plot_columns_distribution(file_path, column_names):
   
    df = pd.read_excel(file_path)

    
    for column_name in column_names:
       
        if column_name in df.columns:
            
            plt.figure()
            
            sns.histplot(df[column_name].dropna(), kde=True) # Drop NA values for clean plotting
            plt.title(f'Distribution of {column_name}')
            plt.xlabel(column_name)
            plt.ylabel('Frequency')
            plt.show()
        else:
            print(f"The column '{column_name}' does not exist in the DataFrame.")


file_path = 'modified_file.xlsx' 
column_names = ['Peak NIR a LogScale15', 'Peak NIR a Log Scale(150)', 'Peak NIR a Log', 'Peak 1 a Log', 'Peak 2 a Log', 'Peak 3 a Log'] 
plot_columns_distribution(file_path, column_names)
```
