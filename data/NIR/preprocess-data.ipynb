```python
def energy_to_wavelength(energy):
    # Constants
    h = 4.135667516E-15  # Planck's constant in eV·s
    c = 299792458  # Speed of light in m/s

    # Convert energy (eV) to wavelength (nm)
    wavelength_nm = 1e9 * h * c / energy
    
    return wavelength_nm

# Example usage
energy_in_ev = 2  # energy in electron volts
wavelength_in_nm = energy_to_wavelength(energy_in_ev)
print(f"{energy_in_ev} electronvolts is {wavelength_in_nm:.1f} nanometers")
```

```python
def wavelength_to_energy(wavelength_nm):
    # Constants
    h = 4.135667516E-15  # Planck's constant in eV·s
    c = 299792458  # Speed of light in m/s

    # Convert wavelength (nm) to energy (eV)
    wavelength_m = wavelength_nm * 1E-9  # Convert nm to meters
    energy_ev = (h * c) / wavelength_m
    
    return energy_ev

# Example usage
wavelength_in_nm = 619.9  # wavelength in nanometers
energy_in_ev = wavelength_to_energy(wavelength_in_nm)
print(f"{wavelength_in_nm} nanometers is {energy_in_ev:.2f} electronvolts")
```

```python
import pandas as pd
import numpy as np

# Define the conversion function
def wavelength_to_energy(wavelength_nm):
    h = 4.135667516E-15  # Planck's constant in eV·s
    c = 299792458  # Speed of light in m/s
    wavelength_m = wavelength_nm * 1e-9  # Convert nm to m
    energy_ev = h * c / wavelength_m
    return energy_ev

# Function to process the Excel file
def process_excel_file(file_path):
    # Read the Excel file
    df = pd.read_excel(file_path)
    
    # Check if "NIR peak" column exists and has values
    if "NIR peak" in df.columns:
        # Convert "NIR peak" values to energy and save in a new column "NIR Energy"
        df["NIR Energy"] = df["NIR peak"].apply(lambda x: wavelength_to_energy(x) if pd.notnull(x) else None)
        
        # For rows where "NIR peak" has a value, set "NIR c" column to sqrt(2)*100
        df["NIR c"] = np.where(pd.notnull(df["NIR peak"]), np.sqrt(2)*100, None)
    
    # Save the updated DataFrame to a new Excel file (or overwrite the existing one)
    df.to_excel("NIR/processed_file.xlsx", index=False)

# Example usage
file_path = "NIR/20240202_gaussian_fits_data_updated.xlsx"
process_excel_file(file_path)
```

```python
import pandas as pd
import numpy as np

def energy_to_wavelength(energy):
    if pd.isnull(energy) or energy == 0:  # Check if the value is NaN, None, or zero
        return None
    else:
        # Constants
        h = 4.135667516E-15  # Planck's constant in eV·s
        c = 299792458  # Speed of light in m/s
        try:
            wavelength_nm = 1e9 * h * c / energy
            return wavelength_nm
        except ZeroDivisionError:
            return None

# Replace 'your_file.xlsx' with the path to your Excel file
file_path = 'NIR/processed_file.xlsx'
output_file_path = 'NIR/Preprocessed_gaussian_fits_data.xlsx'

# Read the Excel file
df = pd.read_excel(file_path)

# Columns for energy to wavelength conversion
energy_columns = ['Peak 1 b', 'Peak 2 b', 'Peak 3 b']
# Columns for log10 calculation
log_columns = ['Peak 1 a', 'Peak 2 a', 'Peak 3 a', 'NIR a']

# Convert energy to wavelength
for column in energy_columns:
    if column in df.columns:
        wav_column_name = f"{column} WAV"
        df[wav_column_name] = df[column].apply(energy_to_wavelength)

# Calculate log10
for column in log_columns:
    if column in df.columns:
        log_column_name = f"{column} Log"
        # Apply log10, handling NaN values appropriately
        df[log_column_name] = df[column].apply(lambda x: np.log10(x) if pd.notnull(x) and x > 0 else None)

# Save the DataFrame with the new columns back to an Excel file
df.to_excel(output_file_path, index=False)

print("The file has been updated with new columns for wavelength values and log10 calculations.")
```

```python
import pandas as pd

# Read the 20240202_gaussian_fits_data.xlsx and NIR-secondary.xlsx files
df_gaussian = pd.read_excel("NIR/20240202_gaussian_fits_data.xlsx")
df_nir_secondary = pd.read_excel("NIR/NIR-secondary.xlsx")

# Merge the two DataFrames based on the "Sequence" column from NIR-secondary
merged_df = pd.merge(df_gaussian, df_nir_secondary, on="Sequence", how="left")

# Filter out rows where "NIR Int Scaled" and "NIR peak" are NaN
filtered_df = merged_df.dropna(subset=["NIR Int Scaled", "NIR peak"], how="any")

# Add "NIR Int Scaled" and "NIR peak" columns to the 20240202_gaussian_fits_data.xlsx file
for index, row in filtered_df.iterrows():
    sequence = row["Sequence"]
    gaussian_index = df_gaussian[df_gaussian["Sequence"] == sequence].index
    df_gaussian.loc[gaussian_index, "NIR Int Scaled"] = row["NIR Int Scaled"]
    df_gaussian.loc[gaussian_index, "NIR peak"] = row["NIR peak"]

# Print sequences in NIR-secondary.xlsx but not in 20240202_gaussian_fits_data.xlsx
missing_sequences = df_nir_secondary[~df_nir_secondary["Sequence"].isin(df_gaussian["Sequence"])]
print("Sequences in NIR-secondary.xlsx but not in 20240202_gaussian_fits_data.xlsx:")
print(missing_sequences["Sequence"].tolist())

# Save the updated 20240202_gaussian_fits_data.xlsx file
df_gaussian.to_excel("NIR/20240202_gaussian_fits_data_updated.xlsx", index=False)
```
